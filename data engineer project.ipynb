```python
import datetime
import random
import string
import pandas as pd
import os
import time
```

```python
letters = string.ascii_letters
patients = [''.join(random.choice(letters) for _ in range(random.randint(0,10))) for _ in range(0,1000)]
```

```python
# Generate input CSV files of random data
def generate_random_csv(patients=patients):
    data = list()
    letters = string.ascii_letters
    for x in range(0,100):
        patient = random.choice(patients)
        description = ''.join(random.choice(letters) for _ in range(random.randint(1,20)))
        timestamp = datetime.datetime.now()
        data.append([patient, description, timestamp])
    df=pd.DataFrame(data)
    df.columns=['id','description','timestamp']
    df.to_csv("input/%s.csv" % ''.join(random.choice(letters) for _ in range(random.randint(10,20))), index=False)
    return df
dfs=[generate_random_csv(patients=patients) for _ in range(0,100)]
```

```python
def read_data():
    files = os.listdir('input')
    dfs = [pd.read_csv('input/'+file) for file in files]
    for file, df in zip(files,dfs):
        df['source_filename'] = file
    df = pd.concat(dfs).reset_index(drop=True)
    final = df.sort_values('timestamp').drop_duplicates('id', keep='last')
    final.to_csv('output/id_descriptions.csv')
    return final
read_data()
```

The first function generates a random file with the columns ['id','description','timestamp'].<br />
It stores it in the output folder.<br />
The second function reads every CSV in the input folder.<br />
It combines them into one dataframe.<br />
It keeps only the latest instance of each id.<br />
It writes the result as a CSV to the output folder.<br />
